```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
```

```python
df_data_2d = pd.read_csv('data/Hw6_data/data_2d.csv')
df_mnist_data = pd.read_csv('data/Hw6_data/mnist.csv')
```

```python
df_data_2d.head()
```

```python
df_mnist_data.head()
```

```python
dist = []
for k in range(1,11):
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(df_data_2d)
    dist.append(kmeanModel.inertia_)
plt.plot(range(1,11), dist, '-r.')
plt.xlabel('k')
plt.show()
```

data_2d - оптимальна кількість кластерів за методом ліктя становить 2 шт.

```python
pca = PCA(n_components = 2)
data_2d = pca.fit_transform(df_data_2d)
kmeans_data_2d = KMeans(n_clusters=2).fit(df_data_2d)
plt.scatter(data_2d[:, 0], data_2d[:, 1], c=kmeans_data_2d.labels_)
```

```python
pca = PCA(n_components = 2)
mnist = pca.fit_transform(df_mnist_data)
```

```python
dist = []
for k in range(1,11):
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(mnist)
    dist.append(kmeanModel.inertia_)
plt.plot(range(1,11), dist, '-r.')
plt.xlabel('k')
plt.show()
```

mnist - оптимальна кількість кластерів за методом ліктя становить 3 шт.

```python
kmeans = KMeans(n_clusters=3).fit(mnist)
plt.scatter(mnist[:, 0], mnist[:, 1], c=kmeans.labels_)
```
